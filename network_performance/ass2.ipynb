```python
import numpy as np
import matplotlib.pyplot as plt


def generateplot(points, label=False):
    xs = [p[0] for p in points]
    ys = [p[1] for p in points]
    if label:
        for p in points:
            plt.text(p[0],p[1],str(p),color='b',fontsize=8)

    plt.grid(True)

    plt.plot(xs, ys, )
    


```

```python
# 2
p = generateplot([(0,0),(10,10),(12.5,0),(15,0),(20,20),(25,0)],True)
plt.title('Water level')
plt.xlabel('time(s)')
plt.ylabel('bufferSize(Mbit)')

plt.show()
```
